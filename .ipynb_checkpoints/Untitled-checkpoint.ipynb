```python
new_df['bhk'] = new_df['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
new_df["total_sqft"].unique()
```

```python
new_df[new_df['total_sqft'].str.contains('-')].head(10)
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
new_df[~new_df["total_sqft"].apply(is_float)].head(10)
```

```python
def convert_sqft(x):
    tokens = x.split('-')
    
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft("2100 - 2850")
```

```python
df2 = new_df.copy()
df2['total_sqft'] = df2['total_sqft'].apply(convert_sqft)
```

```python
df2.loc[122]
```

```python
df3 = df2.copy()
df3["price_per_sqft"] = df3["price"]*100000 / df3["total_sqft"]
df3.head()
```
